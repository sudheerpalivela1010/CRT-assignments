```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def generate_wordcloud(text, output_file):
    """
    Generates a WordCloud from the given text and saves it as an image.
    
    Args:
    text (str): The input text to generate the WordCloud.
    output_file (str): The filename to save the WordCloud image.
    """

    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')  # Turn off the axis
    plt.title("WordCloud Visualization", fontsize=16)
    plt.show()

    wordcloud.to_file(output_file)
    print(f"WordCloud saved as {output_file}")

input_text = "data science machine learning artificial intelligence"
output_image = "wordcloud.png"
generate_wordcloud(input_text, output_image)
```
